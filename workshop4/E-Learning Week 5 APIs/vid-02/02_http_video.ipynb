```python
import requests
```

### Example 1

```python
# Request 1: Random Joke
response_joke = requests.get('https://official-joke-api.appspot.com/random_joke')
if response_joke.status_code == 200:
    joke = response_joke.json()
    print("Joke:")
    print(f"""
    Q: {joke['setup']}
    A: {joke['punchline']}""")
else:
    print(f"Joke API request failed with status code: {response_joke.status_code}")
```

### Example 2a

```python
# Request 2: 24-hour weather forecast from data.gov.sg
response_forecast = requests.get('https://api-open.data.gov.sg/v2/real-time/api/twenty-four-hr-forecast')
if response_forecast.status_code == 200:
    region = 'central'
    print(f"Forecast for {region} region:")
    for fc in response_forecast.json()['data']['records'][0]['periods']:
        print(f"* {fc['timePeriod']['text']}:\t {fc['regions'][region]['text']}")    
else:
    print(f"Forecast retrieval failed: {response_forecast.status_code}")
```

### Example 2b

```python
five_min_url = "https://api-open.data.gov.sg/v2/real-time/api/rainfall"
params = {'date' : '2025-02-01T11:05:00'}
five_min_out = requests.get(five_min_url, params=params)
```

### Example 3

```python
# Request 3: Daily weather data from weather.gov.sg
url_daily_weather = 'https://www.weather.gov.sg/files/dailydata/DAILYDATA_S24_202502.csv'
response_daily_weather = requests.get(url_daily_weather)
```

```python
if response_daily_weather.status_code == 200:
    with open('DAILYDATA_S24_202502.csv', 'wb') as file:
        file_size = file.write(response_daily_weather.content)
    print(f"Downloaded DAILYDATA_S24_202502.csv, size: {len(response_daily_weather.content)} bytes")
else:
    print(f"Daily weather data request failed: {response_daily_weather.status_code}")
```
