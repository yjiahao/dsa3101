```python
import pandas as pd
import folium
import pyproj

df = pd.read_csv('HDBCarparkInformation.csv')
```

Transform coordinates from SVY21 projection to Lat/Long:

```python
transformer = pyproj.Transformer.from_crs("EPSG:3414", "EPSG:4326", always_xy=True)
lat, long = transformer.transform(df.x_coord, df.y_coord)

df['Latitude'] = lat
df['Longitude'] = long
```

```python
# Create a folium map centered around Singapore
sg_map = folium.Map(location=[1.3521, 103.8198], zoom_start=12)

for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Longitude'], row['Latitude']],
        radius=2,
        color='blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(sg_map)
```
